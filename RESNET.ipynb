```
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
%matplotlib inline

from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("MNIST_data", one_hot=True)
```

```
train_img=mnist.train.images
train_label=mnist.train.labels
test_img=mnist.test.images
test_label=mnist.test.labels
```

```
def build_model(x):
  conv1_1=tf.layers.conv2d(inputs=x,filters=32,kernel_size=[3,3],padding='SAME',activation=tf.nn.relu,kernel_initializer=tf.initializers.glorot_normal())
  conv1_2=tf.layers.conv2d(inputs=conv1_1,filters=32,kernel_size=[3,3],padding='SAME',activation=tf.nn.relu,kernel_initializer=tf.initializers.glorot_normal())
  maxp1=tf.layers.max_pooling2d(inputs=conv1_2+x,pool_size=[2,2],strides=(2,2)) # 14 14 32
  
  conv2_1=tf.layers.conv2d(inputs=maxp1,filters=64,kernel_size=[3,3],padding='SAME',activation=tf.nn.relu,kernel_initializer=tf.initializers.glorot_normal())
  conv2_2=tf.layers.conv2d(inputs=conv2_1,filters=64,kernel_size=[3,3],padding='SAME',activation=tf.nn.relu,kernel_initializer=tf.initializers.glorot_normal())
  maxp2=tf.layers.max_pooling2d(inputs=conv2_2+conv2_1,pool_size=[2,2],strides=(2,2)) # 7 7 64
  
  #flatten=tf.reshape(maxp2,[-1,7*7*64])
  flatten=tf.layers.flatten(maxp2)
  hidden1=tf.layers.dense(inputs=flatten,units=100,activation=tf.nn.relu,kernel_initializer=tf.initializers.glorot_normal())
  output=tf.layers.dense(inputs=hidden1,units=10,kernel_initializer=tf.initializers.glorot_normal())
  return output
  
```

```
x=tf.placeholder(tf.float32,[None,28,28,1])
y=tf.placeholder(tf.float32,[None,10])
```

```
pred=build_model(x)
loss=tf.nn.sigmoid_cross_entropy_with_logits(logits=pred,labels=y)
loss=tf.reduce_mean(loss)
opt=tf.train.AdamOptimizer(0.0001).minimize(loss)
sess=tf.Session()
sess.run(tf.global_variables_initializer())
```

```
def get_batch(image,label,batch_size=64):
  idx=np.random.choice(image.shape[0],batch_size)
  return image[idx],label[idx]
```

```
for i in range(1000):
  train_x,train_y=get_batch(train_img.reshape(-1,28,28,1),train_label)
  _,l=sess.run([opt,loss],feed_dict={x:train_x,y:train_y})
  if i%100==0:
    print(l)
```

```
test_pred=sess.run(tf.nn.softmax(pred),feed_dict={x:test_img.reshape(-1,28,28,1)})
print(test_pred.shape)
test_pred=np.argmax(test_pred,axis=1)
print(test_pred.shape)
test_label=np.argmax(test_label,axis=1)
print(np.mean(np.equal(test_pred,test_label)))
```
