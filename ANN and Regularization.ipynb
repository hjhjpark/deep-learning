```python

import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python

N = 10
data_x = np.linspace(-4.5, 4.5, N)
data_y = np.array([0.9819, 0.7973, 1.9737, 0.1838, 1.3180, -0.8361, -0.6591, -2.4701, -2.8122, -6.2512])

data_x = data_x.reshape(-1,1)
data_y = data_y.reshape(-1,1)

plt.figure(figsize = (10,8))
plt.plot(data_x, data_y, 'o')
plt.grid(alpha = 0.3)
plt.show()
```

```python

n_input = 1
n_hidden1 = 30
n_hidden2 = 100
n_hidden3 = 100
n_hidden4 = 30
n_output = 1
```

```python
weights={
    'hidden1':tf.Variable(tf.random_normal([n_input,n_hidden1],stddev=0.1),dtype=tf.float32),
        'hidden2':tf.Variable(tf.random_normal([n_hidden1,n_hidden2],stddev=0.1),dtype=tf.float32),
        'hidden3':tf.Variable(tf.random_normal([n_hidden2,n_hidden3],stddev=0.1),dtype=tf.float32),
        'hidden4':tf.Variable(tf.random_normal([n_hidden3,n_hidden4],stddev=0.1),dtype=tf.float32),
        'output':tf.Variable(tf.random_normal([n_hidden4,n_output],stddev=0.1),dtype=tf.float32)

}
bias={
    'hidden1':tf.Variable(tf.random_normal([n_hidden1],stddev=0.1),dtype=tf.float32),
    'hidden2':tf.Variable(tf.random_normal([n_hidden2],stddev=0.1),dtype=tf.float32),
    'hidden3':tf.Variable(tf.random_normal([n_hidden3],stddev=0.1),dtype=tf.float32),
    'hidden4':tf.Variable(tf.random_normal([n_hidden4],stddev=0.1),dtype=tf.float32),
    'output':tf.Variable(tf.random_normal([n_output],stddev=0.1),dtype=tf.float32),
}
```

```python
def build_model(x,weights,bias):
    hidden1=tf.add(tf.matmul(x,weights['hidden1']),bias['hidden1'])
    hidden1=tf.nn.sigmoid(hidden1)
    
    hidden2=tf.add(tf.matmul(hidden1,weights['hidden2']),bias['hidden2'])
    hidden2=tf.nn.sigmoid(hidden2)
    
    hidden3=tf.add(tf.matmul(hidden2,weights['hidden3']),bias['hidden3'])
    hidden3=tf.nn.sigmoid(hidden3)
    
    hidden4=tf.add(tf.matmul(hidden3,weights['hidden4']),bias['hidden4'])
    hidden4=tf.nn.sigmoid(hidden4)
    
    output=tf.add(tf.matmul(hidden4,weights['output']),bias['output'])
    
    return output
```

```python
x=tf.placeholder(tf.float32,[None,n_input])
y=tf.placeholder(tf.float32,[None,n_output])

pred=build_model(x,weights,bias)
loss=tf.square(pred-y)
loss=tf.reduce_mean(loss)
opt=tf.train.GradientDescentOptimizer(0.01).minimize(loss)
init=tf.global_variables_initializer()

sess=tf.Session()
sess.run(init)
for i in range(100000):
    _,l=sess.run([opt,loss],feed_dict={x:data_x,y:data_y})
    print(l)
```

```python
xp=np.linspace(-4.5,4.5,100).reshape(-1,1)
new_pred=sess.run(pred,feed_dict={x:xp})
```

```python
plt.figure(figsize = (10,8))
plt.plot(data_x, data_y, 'o')
plt.plot(xp, new_pred, 'r')
plt.grid(alpha = 0.3)
plt.show()
```

```python
x=tf.placeholder(tf.float32,[None,n_input])
y=tf.placeholder(tf.float32,[None,n_output])
lamb=0.01
pred = build_model(x, weights, bias)
regularizer = tf.norm(weights['hidden1'], ord = 2) + \
              tf.norm(weights['hidden2'], ord = 2) + \
              tf.norm(weights['hidden3'], ord = 2) + \
              tf.norm(weights['hidden4'], ord = 2) + \
              tf.norm(weights['output'], ord = 2)

loss=tf.square(pred-y)
loss=tf.reduce_mean(loss)+lamb*regularizer

optm = tf.train.AdamOptimizer(0.0001).minimize(loss)


n_batch = 50    
n_iter = 10000  
n_prt = 1000    

sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)

loss_record = []

for epoch in range(n_iter):
    idx = np.random.randint(N, size = n_batch)
    train_x = data_x[idx,:]
    train_y = data_y[idx,:]
    
    sess.run(optm, feed_dict = {x: data_x,  y: data_y})
    
    if epoch % n_prt == 0:
        c = sess.run(loss, feed_dict = {x: data_x, y: data_y})
        loss_record.append(c)
        #print ("Iter : {}".format(epoch))
        #print ("Train Cost : {}".format(c))

        
plt.figure(figsize = (10,8))
plt.plot(np.arange(len(loss_record))*n_prt, loss_record, label = 'training')
plt.xlabel('iteration', fontsize = 15)
plt.ylabel('loss', fontsize = 15)
plt.grid('on', alpha = 0.3)
plt.legend(fontsize = 12)
plt.ylim([0, 10])
plt.show()  
```

```python

xp = np.linspace(-4.5, 4.5, 100).reshape(-1,1)
my_pred = sess.run(pred, feed_dict = {x: xp})

plt.figure(figsize = (10,8))
plt.plot(data_x, data_y, 'o')
plt.plot(xp, my_pred, 'r')
plt.grid(alpha = 0.3)
plt.show()
```

```python
p = tf.placeholder(tf.float32)
x = tf.placeholder(tf.float32, [None, n_input])
y = tf.placeholder(tf.float32, [None, n_output])
```

```python
def build_model(x,weights,bias,p):
    hidden1=tf.add(tf.matmul(x,weights['hidden1']),bias['hidden1'])
    hidden1=tf.nn.sigmoid(hidden1)
    dropout1=tf.nn.dropout(hidden1,keep_prob=p)
    
    hidden2=tf.add(tf.matmul(hidden1,weights['hidden2']),bias['hidden2'])
    hidden2=tf.nn.sigmoid(hidden2)
    dropout2=tf.nn.dropout(hidden2,keep_prob=p)
    
    hidden3=tf.add(tf.matmul(hidden2,weights['hidden3']),bias['hidden3'])
    hidden3=tf.nn.sigmoid(hidden3)
    dropout3=tf.nn.dropout(hidden3,keep_prob=p)
    
    hidden4=tf.add(tf.matmul(hidden3,weights['hidden4']),bias['hidden4'])
    hidden4=tf.nn.sigmoid(hidden4)
    dropout4=tf.nn.dropout(hidden4,keep_prob=p)
    
    output=tf.add(tf.matmul(hidden4,weights['output']),bias['output'])
    
    return output
```

```python

pred = build_model(x, weights, bias, p)
loss = tf.square(pred - y)
loss = tf.reduce_mean(loss)

LR = 0.001
optm = tf.train.AdamOptimizer(LR).minimize(loss)
```

```python

n_batch = 50 
n_iter = 10000
n_prt = 1000  

sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)

loss_record = []
for epoch in range(n_iter):
    idx = np.random.randint(N, size = n_batch)
    train_x = data_x[idx,:]
    train_y = data_y[idx,:]
    
    sess.run(optm, feed_dict = {x: train_x,  y: train_y, p: 0.2})
    
    if epoch % n_prt == 0:
        c = sess.run(loss, feed_dict = {x: train_x, y: train_y, p: 0.2})
        loss_record.append(c)
```

```python
plt.figure(figsize = (10,8))
plt.plot(np.arange(len(loss_record))*n_prt, loss_record, label = 'training')
plt.xlabel('iteration', fontsize = 15)
plt.ylabel('loss', fontsize = 15)
plt.grid('on', alpha = 0.3)
plt.legend(fontsize = 12)
plt.ylim([0, 10])
plt.show()   
```
