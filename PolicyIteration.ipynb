```python
import numpy as np

P = {    
    0: {0: [(1, 0)], 
        1: [(0.5, 0), (0.5, 1)]},
    1: {0: [(0.5, 0), (0.5, 3)], 
        1: [(1, 1)]},
    2: {0: [(0.5, 0), (0.5, 2)], 
        1: [(0.5, 0), (0.5, 1)]},
    3: {0: [(0.5, 2), (0.5, 3)], 
        1: [(1, 1)]},
}

R = [0, 0, 10, 10]
gamma = 0.9

States = [0, 1, 2, 3]
Actions = [0, 1]

v = [0]*4
optPolicy = [0]*4
```

```python

# shorten

v = [0]*4

for j in range(100):
    for i in range(10):
        for s in States:
            v[s] = R[s] + gamma*[sum([trans[0]*v[trans[1]] for trans in P[s][a]]) for a in Actions][optPolicy[s]]
    
    for s in States:
        optPolicy[s]=np.argmax([sum([trans[0]*v[trans[1]] for trans in P[s][a]]) for a in Actions])
    
    
print(v)
print(optPolicy)
```
